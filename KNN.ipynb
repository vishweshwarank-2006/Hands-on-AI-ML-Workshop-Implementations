```python
x = data.drop(['Id', 'Species'], axis = 1)
```

```python
x.shape
```

```python
x.head()
```

```python
y = data['Species']
# 2 brackets (()), it is dataframe
```

```python
# to check value_counts, it shd be either in panda series or dataframe
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.8, random_state = 3)
# for splitting the data into training and testing data by 80% and 20%
```

```python
print (xtrain.shape)
print (xtest.shape)
print (ytrain.shape)
print (ytest.shape)
```

```python
# now we have split 80% of y in each 50, 80% of 5 for all 3
ytrain.value_counts()
# to get 40, 40, 40, tune the model by changing random state, for this, it shd be 3
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ytest.values
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
ac = accuracy_score(ytest, ypred)
ac
```
