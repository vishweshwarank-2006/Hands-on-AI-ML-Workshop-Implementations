```python
#Classification problem using DL
import tensorflow
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import*
```

```python
data = pd.read_csv('/content/breast_cancer _ day1.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.drop(['id'], axis= 1)
```

```python
data.shape
```

```python
#earlier we had 32,now we removed id, and it became 31
# x = must have 31 features, y = cancer or not, y must have diagnosis column
x = data.drop(['diagnosis'], axis =1)
```

```python
x.shape
```

```python
# x has 569 entries, every entry has 30 features
y = data['diagnosis'] #1-D, shd have not two brackets '[[]]'
```

```python
y.shape
```

```python
#we use standard scaler function because we have different data types thats why
#and also we are converting the label encounter to integer
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y=pd.DataFrame(y)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
x = pd.DataFrame(x)
```

```python
# x has been converted to dataframe
x.head()
```

```python
from sklearn.model_selection import (train_test_split)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size= 0.8, random_state= 0)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# y is in panda series (binary)
y.value_counts()
#not the total 80% of 569 we need 80% of both b and m values respectively so that the split what we do should be correct
```

```python
ytrain.value_counts()
```

```python
model = Sequential()
```

```python
model.add(Dense(16,input_dim=30,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(xtrain,ytrain,epochs=50,validation_data=(xtest,ytest))
```

```python
ypred = model.predict(xtest)
```

```python
ypred
ypred = ypred.round()
```
