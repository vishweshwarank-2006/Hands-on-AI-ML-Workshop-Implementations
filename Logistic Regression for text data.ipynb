```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
sentences = ['I love this movie', ' I hate this movie', 'This film is amazing', 'this movie is horrible']
labels = ['positive', 'negative', 'positive', 'negative']
```

```python
tfidf =  TfidfVectorizer()
x = tfidf.fit_transform(sentences).toarray()
y =labels
```

```python
print(x)
```

```python
# model already imported in the beginning
model = LogisticRegression()
model.fit(x,y)
```

```python
# y is in list, it can be 1-D array, panda series or list
# sk learn - no need to convert y to number (now it is in text, ml can accomodate that, DL will give an error)

new_sentence = ['I personally love the film a lot']
```

```python
x_new = tfidf.transform(new_sentence).toarray()
prediction = model.predict(x_new)
```
